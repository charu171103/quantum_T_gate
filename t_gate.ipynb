```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def initialize_state(circuit, state):
    if state == '|0>':
        pass  
    elif state == '|1>':
        circuit.x(0)  
    elif state == '|+>':
        circuit.h(0) 
    elif state == '|->':
        circuit.h(0)
        circuit.z(0)  
    elif state == '|i>':
        circuit.h(0)
        circuit.s(0) 
    elif state == '|-i>':
        circuit.h(0)
        circuit.sdg(0) 
```

```python
def simulate_t_gate(input_state):
    qc = QuantumCircuit(1)

    initialize_state(qc, input_state)

    # Apply the T gate
    qc.t(0)

    # Get the statevector
    statevector = Statevector.from_instruction(qc)

    print(f"Input State: {input_state}")
    print(f"Output Statevector: {statevector.data}\n")
    
    print("Q-Sphere Visualization:")
    display(plot_state_qsphere(statevector))

    print("Bloch Sphere Visualization:")
    display(plot_bloch_multivector(statevector))

    print("Circuit:")
    display(qc.draw(output='mpl'))
    print("-" * 50)
```

```python
def main():
    input_states = ['|0>', '|1>', '|+>', '|->', '|i>', '|-i>']
    for input_state in input_states:
        simulate_t_gate(input_state)
```

```python
if __name__ == "__main__":
    main()
```
